```python
import os
import pdfplumber
import docx
import pandas as pd
import re
import win32com.client as win32
import textract
import subprocess
import time

# Function to extract text from PDF
# def extract_text_from_pdf(pdf_path):
#     with pdfplumber.open(pdf_path) as pdf:
#         full_text = ""
#         for page in pdf.pages:
#             full_text += page.extract_text() or ""
#     return full_text

def extract_text_from_docx(docx_path):
    retry_attempts = 3
    for attempt in range(retry_attempts):
        try:
            doc = docx.Document(docx_path)
            full_text = [para.text for para in doc.paragraphs if para.text]
            return '\n'.join(full_text)
        except Exception as e:
            print(f"Attempt {attempt + 1}: Unable to read {docx_path}, error: {e}")
            time.sleep(2)  # wait for 2 seconds before retrying
    return None


# Function to extract text from DOCX
def extract_text_from_docx(docx_path):
    doc = docx.Document(docx_path)
    full_text = [para.text for para in doc.paragraphs if para.text]
    return '\n'.join(full_text)

# def convert_doc_to_docx(doc_path, output_dir):
#     if not os.path.exists(output_dir):
#         os.makedirs(output_dir)
#     output_path = os.path.join(output_dir, os.path.splitext(os.path.basename(doc_path))[0] + '.docx')
#     # Ensure the path is enclosed in quotes in case there are spaces
#     # command = f'libreoffice --headless --convert-to docx:writer_pdf_Export --outdir "{output_dir}" "{doc_path}"'
#     command = f'"C:\\Program Files (x86)\\LibreOffice\\program\\soffice" --headless --convert-to docx:writer_pdf_Export --outdir "{output_dir}" "{doc_path}"'

#     # Run the command and capture output and errors
#     result = subprocess.run(command, shell=True, capture_output=True, text=True)
#     if result.returncode != 0:
#         print(f"Error converting file {doc_path}: {result.stdout} {result.stderr}")
#     return output_path

# def convert_doc_to_docx(doc_path, output_dir):
#     if not os.path.exists(output_dir):
#         print(f"Creating output directory at {output_dir}")
#         os.makedirs(output_dir)
#     output_path = os.path.join(output_dir, os.path.splitext(os.path.basename(doc_path))[0] + '.docx')
#     command = f'"C:\\Program Files (x86)\\LibreOffice\\program\\soffice" --headless --convert-to docx:writer_pdf_Export --outdir "{output_dir}" "{doc_path}"'
#     print(f"Running command: {command}")
#     result = subprocess.run(command, shell=True, capture_output=True, text=True)
#     if result.returncode != 0:
#         print(f"Error converting file {doc_path}: {result.stdout} {result.stderr}")
#     else:
#         print(f"Conversion successful: {output_path}")
#     return output_path

# def convert_doc_to_docx(doc_path, output_dir):
#     if not os.path.exists(output_dir):
#         os.makedirs(output_dir)
#     output_path = os.path.join(output_dir, os.path.splitext(os.path.basename(doc_path))[0] + '.docx')
#     command = f'"C:\\Program Files (x86)\\LibreOffice\\program\\soffice" --headless --convert-to docx:writer_pdf_Export --outdir "{output_dir}" "{doc_path}"'
#     result = subprocess.run(command, shell=True, capture_output=True, text=True)
#     if result.returncode != 0:
#         print(f"Error converting file {doc_path}: {result.stdout} {result.stderr}")
#     else:
#         print(f"Conversion successful: {output_path}")
#     if os.path.exists(output_path):
#         print(f"File {output_path} exists.")
#     else:
#         print(f"File {output_path} does not exist.")
#     return output_path
def convert_doc_to_docx(doc_path, output_dir):
    output_path = os.path.join(output_dir, os.path.splitext(os.path.basename(doc_path))[0] + '.docx')
    command = f'"C:\\Program Files (x86)\\LibreOffice\\program\\soffice" --headless --convert-to docx:writer_pdf_Export --outdir "{output_dir}" "{doc_path}"'
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    print(f"Command executed with exit code {result.returncode}")
    if result.returncode != 0:
        print(f"Error: {result.stderr}")
    return output_path


# Regex functions for extracting emails and phone numbers
def extract_email(text):
    email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    return re.findall(email_regex, text)

def extract_phone_number(text):
    phone_regex = r'\b\d{3}[-.]?\d{3}[-.]?\d{4}\b'
    return re.findall(phone_regex, text)

# Processing all files in the directory
def process_files(directory):
    data = []
    output_dir = os.path.join(directory, 'converted')
    print(f"Processing files in {directory}")
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        print(f"Processing file: {filename}")
        if filename.lower().endswith(".pdf"):
            text = extract_text_from_pdf(os.path.join(directory, filename))
        elif filename.lower().endswith(".docx"):
            text = extract_text_from_docx(os.path.join(directory, filename))
        elif filename.lower().endswith(".doc"):
            docx_path = convert_doc_to_docx(file_path, output_dir)
            if os.path.exists(docx_path):
                file_size = os.path.getsize(docx_path)
                print(f"File size of {docx_path} is {file_size} bytes.")
                if file_size > 0:
                    text = extract_text_from_docx(docx_path)
                else:
                    print("Conversion resulted in an empty file.")
        else:
            print(f"Skipping unsupported file format: {filename}")
            continue
        
        if text:
            emails = extract_email(text)
            phones = extract_phone_number(text)
            data.append({"Filename": filename, "Emails": emails, "Phone Numbers": phones, "Text": text})
        else:
            print(f"No text found in {filename}")
    return data

# Write data to Excel
def create_excel(data, filename='output.xlsx'):
    df = pd.DataFrame(data)
    df.to_excel(filename, index=False, engine='openpyxl')

# Directory containing CVs
directory = r'C:\ost_internshala\Sample2'
data = process_files(directory)
create_excel(data)
```
